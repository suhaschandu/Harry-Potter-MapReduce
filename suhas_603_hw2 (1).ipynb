```python
import PyPDF2
from collections import defaultdict
from spellchecker import SpellChecker
```

```python
def extract_pages(pdf_path, start, end, filename):
    pdf = open(pdf_path, "rb")
    reader = PyPDF2.PdfReader(pdf)
    with open(filename, "w", encoding="utf-8") as f:
        for i in range(start-1, end):
            if i < len(reader.pages):
                text = reader.pages[i].extract_text()
                if text:
                    f.write(text + "\n")
    pdf.close()
    print(f" Extracted pages {start}-{end} into {filename}")
```

```python
def word_count(filename):
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()

    words = text.lower().replace("\n", " ").split()
    mapped = [(w.strip(".,!?;:'\"()"), 1) for w in words if w.strip()]

    counts = defaultdict(int)
    for word, c in mapped:
        counts[word] += c

    return counts
```

```python
def non_english_count(filename):
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()

    words = text.lower().replace("\n", " ").split()
    words = [w.strip(".,!?;:'\"()") for w in words if w.strip()]

    spell = SpellChecker()
    non_english = spell.unknown(words)

    mapped = [(w, 1) for w in words if w in non_english]

    counts = defaultdict(int)
    for word, c in mapped:
        counts[word] += c

    return counts
```

```python
#dob-03/14/2002
pdf_path = "/content/harrypotter.pdf"

#my month is march so i have taken the 3rd book in the pdf it starts from 568 and I have taken the 14th page from there
extract_pages(pdf_path, 582, 591, "file1.txt")

#my year is 2002 so i have taken the 3rd book in the pdf it starts from 568 and I have taken the 102th page from there
extract_pages(pdf_path, 668, 679, "file2.txt")
```

```python
wc = word_count("file1.txt")
print("\n Word Count from file1.txt (first 20 words):")
for word, count in list(wc.items()):
    print(word, ":", count)

```

```python
nec = non_english_count("file2.txt")
print("\n Non-English Words from file2.txt:")
for word, count in nec.items():
    print(word, ":", count)
```
